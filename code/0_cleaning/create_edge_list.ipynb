```python
edge_list = df[["investor_permalink", "company_permalink"]]

# First, split the dates by the delimters (either '/' or '-')
temp = df["funded_at"].apply(lambda x: x.split('/') if '/' in x else x.split('-'))

# Convert to integer and put in the correct order (YMD)
temp2 = temp.apply(lambda x: [int(x[2]), int(x[0]), int(x[1])] if int(x[2]) > 1000 else [int(i) for i in x])

# Convert to datetime object
edge_list.loc[:, "invest_date"] = temp2.apply(lambda x: dt.datetime(x[0], x[1], x[2]))
#edge_list.loc[:, "raised_amount_usd"] = df["raised_amount_usd"].apply(lambda x: int(str(x).strip().replace(",", "").replace("-","0")) if not pd.isnull(x) else 0)
edge_list.loc[:, "raised_amount_usd"] = df["raised_amount_usd"].apply(lambda x: int(x) if not pd.isnull(x) else 0)
```

```python
in_degree_hist = Counter(filter(lambda x: x != 0, g2.in_degree().values()))
plt.plot(in_degree_hist.keys(), in_degree_hist.values())
plt.xlabel('In-Degree')
plt.ylabel('Number of Companies')
plt.title("In-Degree Graph: Number of Incoming Investments, 2005-2015")
plt.show()
```

```python
out_degree_hist = Counter(filter(lambda x: x != 0, g2.out_degree().values()))
plt.plot(out_degree_hist.keys(), out_degree_hist.values())
plt.xlabel('Out-Degree')
plt.ylabel('Number of Companies')
plt.title("Out-Degree Graph: Number of Outgoing Investments, 2005-2015")
plt.show()
```

```python
in_degree_hist = Counter(filter(lambda x: x != 0, g2.in_degree().values()))
plt.loglog(in_degree_hist.keys(), in_degree_hist.values())
plt.xlabel('Log In-Degree')
plt.ylabel('Log Number of Companies')
plt.title("In-Degree Graph: Number of Incoming Investments, 2005-2015 (log-log scale)")
plt.show()
```

```python

out_degree_hist = Counter(filter(lambda x: x != 0, g2.out_degree().values()))
plt.loglog(sorted(out_degree_hist.keys()), out_degree_hist.values())
plt.xlabel('Log Out-Degree')
plt.ylabel('Log Number of Companies')
plt.title("Out-Degree Graph: Number of Outgoing Investments, 2005-2015 (log-log scale)")
plt.show()
```
