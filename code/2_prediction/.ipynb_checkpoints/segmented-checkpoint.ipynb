```python
def loadEdgelist(filename):
	# Readin graph
	g = nx.read_edgelist(filename, 
	                     delimiter="\t", 
	                     create_using  = nx.DiGraph(), 
	                     data=True)

	# Convert Weight To Float
	temp = map(lambda (x, y): (x, float(y)), 
			   nx.get_edge_attributes(g, "weight").items())
	nx.set_edge_attributes(g, "weight", dict(temp))

	# Convert Date to datetime
	temp = map(lambda (x, y): (x, dt.datetime.strptime(y, "%Y-%m-%d")), 
			   nx.get_edge_attributes(g, "date").items())
	nx.set_edge_attributes(g, "date", dict(temp))

	return g
```

```python
preG = nx.read_gpickle("preG_pickle_cluster.gpickle")
```

```python
nx.get_node_attributes(preG, "cluster")
```

```python
cluster_attributes = nx.get_node_attributes(preG, "cluster")
indegree_attributes = preG.in_degree()
outdegree_attributes = preG.out_degree()
```

```python
group0 = set(map(lambda x: x[0], 
             filter(lambda x: x[1] == 0, cluster_attributes.items())))
group1 = set(map(lambda x: x[0], 
             filter(lambda x: x[1] == 1, cluster_attributes.items())))
group2 = set(map(lambda x: x[0], 
             filter(lambda x: x[1] == 2, cluster_attributes.items())))

cluster0 = max(map(lambda x: x[1], 
               filter(lambda x: x[0] in group0, 
                      indegree_attributes.items())))
cluster1 = max(map(lambda x: x[1], 
               filter(lambda x: x[0] in group1, 
                      indegree_attributes.items())))

cluster2 = max(map(lambda x: x[1], 
               filter(lambda x: x[0] in group2, 
                      indegree_attributes.items())))

o_cluster0 = max(map(lambda x: x[1], 
               filter(lambda x: x[0] in group0, 
                      outdegree_attributes.items())))
o_cluster1 = max(map(lambda x: x[1], 
               filter(lambda x: x[0] in group1, 
                      outdegree_attributes.items())))

o_cluster2 = max(map(lambda x: x[1], 
               filter(lambda x: x[0] in group2, 
                      outdegree_attributes.items())))

cluster_constant = [0,0,0]
cluster_constant[0] = cluster0 * o_cluster0
cluster_constant[1] = cluster1 * o_cluster1
cluster_constant[2] = cluster2 * o_cluster2
print cluster_constant
print cluster0, cluster1, cluster2
print o_cluster0, o_cluster1, o_cluster2
```
